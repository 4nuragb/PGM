K-Mean Algorithm & Implementation

STEP 1) Data Preprocessing

1) Import the required libraries <br>
2) Importing the dataset 'Mall_Customers.csv'<br>
3) Extracting Independent Variables: Here we don't need any dependent variables for data preprocessing step as it is a clustering problem and we have no idea about what to determine.

```python
import numpy as np
import matplotlib.pyplot as mtp
import pandas as pd
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/4nuragb/PGM/main/Mall_Customers.csv?token=GHSAT0AAAAAABP4ECWZ5OTIIBYVBXMG255OYO5RLGA')
```

```python
x = dataset.iloc[:, [2,4]].values  
```

STEP 2) Finding optimal number of clusters using the Elbow Method
<br>We are going to calculate the WCSS values for different values of k ranging from 1 to 10.

```python
from sklearn.cluster import KMeans 
wcss_list=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)
  wcss_list.append(kmeans.inertia_)
mtp.plot(range(1,11),wcss_list)
mtp.title('The Elbow Method Graph')
mtp.xlabel('Number of Clusters(K)')
mtp.ylabel('wcss_list')
mtp.show()
```

From the above code we have used the sklearn.cluster library to form the clusters. Then we have created a wcss_list variable to initialize an empty list, which is used to contain the WCSS value computed for different values of k ranging from 1 to 10.
After that we have initialized the for loop for the iteration on a different value of k ranging from 1 to 10.
In the next part of code we have fitted he model on a matrix of features and then plotted the graph between number of clusters and WCSS.

STEP 3) Training the K-means algorithm on the training dataset<br>
From STEP 2, we have found the optimal value of k to be 5.<br>
1) First we create the object of KMeans class<br>
2) Then we create the dependent varaible y_predict to train the model.<br>
By executing the above lines of code we will get the y_predict variable.<br>
Now we can compare the values of y_predict with our original dataset.

```python
kmeans= KMeans(n_clusters=4,init='k-means++',random_state=42)
y_predict=kmeans.fit_predict(x)
```

STEP 4) Visualizing the Clusters<br>
To visualize the clusters we will use scatter plot using mtp.scatter() function of matplotlib

```python
mtp.scatter(x[y_predict==0,0],x[y_predict==0,1],s=50,c='blue',label='Cluster1')
mtp.scatter(x[y_predict==1,0],x[y_predict==1,1],s=50,c='red',label='Cluster2')
mtp.scatter(x[y_predict==2,0],x[y_predict==2,1],s=50,c='green',label='Cluster3')
mtp.scatter(x[y_predict==3,0],x[y_predict==3,1],s=50,c='cyan',label='Cluster4')
# mtp.scatter(x[y_predict==4,0],x[y_predict==4,1],s=100,c='magenta',label='Cluster5')

mtp.title('Clusters of customers')
mtp.xlabel('Age(years)')
mtp.ylabel('Spending Score(1-100)')
mtp.legend()
mtp.show()
```
